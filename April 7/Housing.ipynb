```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data_df = pd.read_csv('housing.csv')
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(18,4))

for i in range(len(axs)):
    axs[i].set_ylim([0, 1200000])
# RM vs Price
_ = sns.regplot(x=data_df['RM'], y=data_df['MEDV'], ax=axs[0])
# LSSAT vs Price
_ = sns.regplot(x=data_df['LSTAT'], y=data_df['MEDV'], ax=axs[1])
# PTRATIO vs Price
_ = sns.regplot(x=data_df['PTRATIO'], y=data_df['MEDV'], ax=axs[2])

plt.show()
```
